```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("sleep_mobile_stress_dataset_15000.csv")
```

```python
sns.countplot(x="stress_level",data=df)
plt.title("Stress Level Distribution")
plt.show()
```

```python
sns.scatterplot(x="daily_screen_time_hours",y="sleep_duration_hours",data=df)
plt.title("Screen time vs Sleep time")
plt.show()
```

```python
df.groupby("gender")["daily_screen_time_hours"].mean()
```

```python
sns.boxplot(x="gender",y="daily_screen_time_hours",data=df)
plt.title("Screen time by Gender")
plt.show()
```

```python
sns.scatterplot(
    x="caffeine_intake_cups",
    y="sleep_duration_hours",
    data=df
)

plt.title("Caffeine Intake vs Sleep")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
